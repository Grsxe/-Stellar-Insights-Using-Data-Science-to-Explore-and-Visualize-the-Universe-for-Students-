```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("star_dataset.csv")

# Clean column names
df.columns = df.columns.str.strip()

df.head()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="Temperature (K)",
    y="Luminosity (L/Lo)",
    hue="Spectral Class"
)
plt.title("Temperature vs Luminosity")
plt.show()
```

```python
plt.figure()
sns.scatterplot(
    data=df,
    x="Distance (ly)",
    y="Temperature (K)",
    hue="Spectral Class"
)
plt.title("Distance vs Temperature")
plt.show()
```

```python
print("""
INSIGHTS:
- Hotter stars tend to be more luminous
- Spectral classes group stars by temperature
- Some distant stars are still very bright due to high luminosity
""")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Temperature (K)",
    y="Luminosity (L/Lo)",
    hue="Spectral Class",
    palette="viridis"
)

plt.title("Temperature vs Luminosity of Stars")
plt.xlabel("Temperature (K)")
plt.ylabel("Luminosity (L/Lo)")
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.histplot(df["Temperature (K)"], bins=15, kde=True)

plt.title("Distribution of Star Temperatures")
plt.xlabel("Temperature (K)")
plt.ylabel("Count")

plt.show()
```
